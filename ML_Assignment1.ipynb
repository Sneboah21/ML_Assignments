```python
#Q1.
#a)
import numpy as np
arr = np.array([1, 2, 3, 6, 4, 5])
reversed_arr = arr[::-1]
print("Reversed array: ", reversed_arr)

#b)
array1 = np.array([[1,2,3],[2,4,5],[1,2,3]])
f1 = array1.flatten()
f2 = array1.ravel()
print("Flattened arr: ",f1)
print("flattened arr: ",f2)

#c)
arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[1, 2], [3, 4]])
comparison = np.array_equal(arr1,arr2)
print("Are the two arrays equal? ", comparison)

#d)
x = np.array([1, 2, 3, 4, 5, 1, 2, 1, 1])
y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3])
def most_frequent(arr):
  vals, counts = np.unique(arr, return_counts=True)
  max_count = max(counts)
  most_freq = vals[np.argmax(counts)]
  indices = np.where(arr == most_freq)[0]
  return most_freq, indices
freq_x, indices_x = most_frequent(x)
freq_y, indices_y = most_frequent(y)
print("x: Most frequent =", freq_x, ", Indices =", indices_x)
print("y: Most frequent =", freq_y, ", Indices =", indices_y)

#e)
gfg = np.matrix('4, 1, 9; 12, 3, 1; 4, 5, 6')
sum_all = gfg.sum()
sum_rows = gfg.sum(axis=1)
sum_cols = gfg.sum(axis=0)
print("Sum of all elements:", sum_all)
print("Sum row-wise:", sum_rows)
print("Sum column-wise:", sum_cols)

#f)
n_array = np.array([[55, 25, 15], [30, 44, 2], [11, 45, 77]])
sum_diag = np.trace(n_array)
eig_vals, eig_vecs = np.linalg.eig(n_array)
inverse = np.linalg.inv(n_array)
determinant = np.linalg.det(n_array)
print("Sum of diagonal elements:", sum_diag)
print("Eigenvalues:", eig_vals)
print("Eigenvectors:\n", eig_vecs)
print("Inverse:\n", inverse)
print("Determinant:", determinant)

#g)
#i)
p1 = np.array([[1, 2], [2, 3]])
q1 = np.array([[4, 5], [6, 7]])
product1 = np.dot(p1, q1)
cov1 = np.cov(p1.T, q1.T)
#ii)
p2 = np.array([[1, 2], [2, 3], [4, 5]])
q2 = np.array([[4, 5, 1], [6, 7, 2]])
product2 = np.dot(p2, q2)
cov2 = np.cov(p2.flatten(), q2.flatten())
print("(g.i) Matrix product:\n", product1)
print("(g.i) Covariance:\n", cov1)
print("(g.ii) Matrix product:\n", product2)
print("(g.ii) Covariance:\n", cov2)

#h)
x = np.array([[2, 3, 4], [3, 2, 9]])
y = np.array([[1, 5, 0], [5, 10, 3]])
# Inner product
inner_product = np.inner(x, y)
# Outer product (flatten first)
outer_product = np.outer(x.flatten(), y.flatten())
# Cartesian product
cartesian_product = np.array([[i, j] for i in x.flatten() for j in y.flatten()])
print("Inner product:\n", inner_product)
print("Outer product:\n", outer_product)
print("Cartesian product:\n", cartesian_product)
```

```python
#Q3
import numpy as np

array = np.array([10, 52, 62, 16, 16, 54, 453])
# (i) Sorted array
sorted_array = np.sort(array)
# (ii) Indices of sorted array
sorted_indices = np.argsort(array)
# (iii) 4 smallest elements
smallest_4 = np.sort(array)[:4]
# (iv) 5 largest elements
largest_5 = np.sort(array)[-5:]  # or use: np.partition(array, -5)[-5:]
print("Sorted:", sorted_array)
print("Sorted indices:", sorted_indices)
print("4 smallest:", smallest_4)
print("5 largest:", largest_5)

array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])

# (i) Integer elements only
integers = array[np.floor(array) == array]

# (ii) Float (non-integer) elements only
floats = array[np.floor(array) != array]
print("Integers:", integers)
print("Floats:", floats)
```

```python
#Q4
#a)
from PIL import Image
import numpy as np

def img_to_array(path):
    try:
        # Open image
        img = Image.open(path)

        # Convert to numpy array
        img_array = np.array(img)

        # Check if Grey Scale or RGB
        if len(img_array.shape) == 2:  # Grey scale image (2D array)
            print("Grey scale image detected")
            print(f"Image shape: {img_array.shape}")
            # Save as single text file
            np.savetxt("greyscale_image.txt", img_array, fmt='%d')
            print("Saved as: greyscale_image.txt")

        elif len(img_array.shape) == 3:  # RGB image (3D array)
            print("RGB image detected")
            print(f"Image shape: {img_array.shape}")
            # Save each channel separately
            np.savetxt("rgb_image_r.txt", img_array[:,:,0], fmt='%d')
            np.savetxt("rgb_image_g.txt", img_array[:,:,1], fmt='%d')
            np.savetxt("rgb_image_b.txt", img_array[:,:,2], fmt='%d')
            print("Saved as: rgb_image_r.txt, rgb_image_g.txt, rgb_image_b.txt")

        return img_array

    except FileNotFoundError:
        print(f"Error: File not found at path: {path}")
        return None
    except Exception as e:
        print(f"Error processing image: {e}")
        return None


img_array = img_to_array('/content/drive/MyDrive/floweryay.jpeg')
if img_array is not None:
    print(f"Successfully processed image with shape: {img_array.shape}")
```

```python
#4 b)
def load_saved_file(image_type='rgb'):
    try:
        if image_type == 'greyscale':
            # Load single greyscale file
            loaded_array = np.loadtxt("greyscale_image.txt", dtype=int)
            print(f"Loaded greyscale image with shape: {loaded_array.shape}")

        elif image_type == 'rgb':
            # Load three separate RGB channel files and combine
            r_channel = np.loadtxt("rgb_image_r.txt", dtype=int)
            g_channel = np.loadtxt("rgb_image_g.txt", dtype=int)
            b_channel = np.loadtxt("rgb_image_b.txt", dtype=int)

            # Stack channels to recreate RGB image
            loaded_array = np.stack((r_channel, g_channel, b_channel), axis=-1)
            print(f"Loaded RGB image with shape: {loaded_array.shape}")

        return loaded_array

    except FileNotFoundError as e:
        print(f"Error: Could not find saved file(s): {e}")
        return None
    except Exception as e:
        print(f"Error loading file: {e}")
        return None

# Load the saved file (adjust image_type based on your image)
loaded_image = load_saved_file('rgb')  # or 'greyscale' if your image is greyscale

# Verify the loaded data matches original
if loaded_image is not None:
    print("File successfully loaded into Jupyter notebook!")

    # Optional: Display basic statistics
    print(f"Min value: {loaded_image.min()}")
    print(f"Max value: {loaded_image.max()}")
    print(f"Mean value: {loaded_image.mean():.2f}")
```

```python
# Complete workflow
from PIL import Image
import numpy as np

# Step 1: Read and save image
img_array = img_to_array('/content/drive/MyDrive/floweryay.jpeg')

# Step 2: Load saved file back
if img_array is not None:
    # Determine if RGB or greyscale based on original array
    if len(img_array.shape) == 2:
        loaded_array = load_saved_file('greyscale')
    else:
        loaded_array = load_saved_file('rgb')

    # Step 3: Verify data integrity
    if loaded_array is not None and np.array_equal(img_array, loaded_array):
        print("Data integrity verified: Original and loaded arrays are identical")
    else:
        print("Original and loaded arrays differ")
```
