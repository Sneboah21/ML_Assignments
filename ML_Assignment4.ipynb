Ques 1: Write a Python program to scrape all available books from the website (https://books.toscrape.com/) Books to Scrape – a live site built for practicing scraping (safe, legal, no anti-bot). For each book, extract the following details:
1. Title
2. Price
3. Availability (In stock / Out of stock)
4. Star Rating (One, Two, Three, Four, Five)

Store the scraped results into a Pandas DataFrame and export them to a CSV file named books.csv.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
Base_url = "https://books.toscrape.com/"
titles= []
prices=[]
availabilities = []
ratings = []

for page in range(1,51):
  url=Base_url.format(page)
  response=requests.get(url)
  soup=BeautifulSoup(response.text,"html.parser")
  books=soup.find_all("article",class_="product_pod")
  for book in books:
    title=book.h3.a["title"]
    titles.append(title)
    price=book.find("p",class_="price_color").text
    prices.append(price)
    availability=book.find("p",class_="instock availability").text.strip()
    availabilities.append(availability)
    rating=book.find("p",class_="star-rating")["class"][1]
    ratings.append(rating)
print("data scrapped sucessfully!")
df = pd.DataFrame({
    "Title": titles,
    "Price": prices,
    "Availability": availabilities,
    "Star Rating": ratings
})
df.to_csv("books.csv", index=False)
print("data saved to books.csv")
```

Q2. Write a Python program to scrape the IMDB Top 250 Movies list (https://www.imdb.com/chart/top/) . For each movie, extract the following details:
1. Rank (1–250)
2. Movie Title
3. Year of Release
4. IMDB Rating

Store the results in a Pandas DataFrame and export it to a CSV file named imdb_top250.csv.

```python
# Install dependencies
!pip install selenium webdriver-manager pandas
!apt-get update
!apt install -y chromium-browser chromium-chromedriver

# Imports
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

# Setup Chrome for Colab
options = webdriver.ChromeOptions()
options.add_argument("--headless")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# -----------------------------------------------
# Q2: Scraping IMDB Top 250 Movies
# -----------------------------------------------
url_imdb = "https://www.imdb.com/chart/top/"
driver.get(url_imdb)

movies = driver.find_elements(By.XPATH, '//li[@class="ipc-metadata-list-summary-item"]')

rank, title, year, rating = [], [], [], []

for i, movie in enumerate(movies, start=1):
    try:
        movie_title = movie.find_element(By.XPATH, './/h3').text
        movie_year = movie.find_element(By.XPATH, './/span[contains(@class,"title")]').text.strip("()")
        movie_rating = movie.find_element(By.XPATH, './/span[contains(@class,"rating")]').text

        rank.append(i)
        title.append(movie_title)
        year.append(movie_year)
        rating.append(movie_rating)
    except:
        continue

df_imdb = pd.DataFrame({
    "Rank": rank,
    "Movie Title": title,
    "Year of Release": year,
    "IMDB Rating": rating
})
df_imdb.to_csv("imdb_top250.csv", index=False)

print("✅ IMDB Top 250 Movies scraped. Saved as imdb_top250.csv")
display(df_imdb.head())

# -----------------------------------------------
# Q3: Scraping Weather Info for Top Cities
# -----------------------------------------------
url_weather = "https://www.timeanddate.com/weather/"
driver.get(url_weather)

cities = driver.find_elements(By.XPATH, '//table[contains(@class,"zebra tb-wt fw va-m")]//tr')

city_name, temperature, condition = [], [], []

for city in cities[1:]:  # skip header row
    try:
        name = city.find_element(By.XPATH, './td[1]//a').text
        temp = city.find_element(By.XPATH, './td[2]').text
        cond = city.find_element(By.XPATH, './td[3]').text

        city_name.append(name)
        temperature.append(temp)
        condition.append(cond)
    except:
        continue

df_weather = pd.DataFrame({
    "City Name": city_name,
    "Temperature": temperature,
    "Weather Condition": condition
})
df_weather.to_csv("weather.csv", index=False)

print("✅ Weather data scraped. Saved as weather.csv")
display(df_weather.head())

# Close browser
driver.quit()
```
